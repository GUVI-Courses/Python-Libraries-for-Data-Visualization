### Step1: Import Libraries

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import datetime as dt
```

### Step2: Create Sample Data

```python
# Generate 10 days of data:
dates = [dt.date(2023, 1, d) for d in range(1, 11)]
sales = [100, 120, 90, 110, 150, 170, 160, 180, 200, 210]
```

### Step3: Basic time series plot

```python
plt.plot(dates, sales, marker = 'o')
plt.title("Daily Sales Over Time")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.show()
```

### Step4: Date Formatting

```python
fig, ax = plt.subplots()

ax.plot(dates, sales, marker = 'o', color = 'green')
ax.set_title("Daily Sales Over Time")

# Format the x-axis:
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b %d')) # Jan 01
plt.xticks(rotation = 45)

plt.xlabel("Date")
plt.ylabel("Sales")

plt.show()
```

### Step 5: Highlighting ranges & adding events:

```python
fig, ax = plt.subplots()

ax.plot(dates, sales, marker = 'o', color = 'blue')
ax.set_title("Sales Trend with Highlighted Period")

# Highlight a range (say, days 4-6)
ax.axvspan(dates[3], dates[5], color = "yellow", alpha = 0.3, label = "Promo Period")

# Mark a specific event:
ax.axvline(dates[7], color = "red", linestyle = '--', label = "New Product Launch")

plt.xticks(rotation = 45)
plt.xlabel("Date")
plt.ylabel("Sales")
plt.legend()
plt.show()
```

### Demo1: Simple Twin Axes

```python
import matplotlib.pyplot as plt
```

```python
# Sample data:
days = [1, 2, 3, 4, 5, 6, 7]
temperature = [22, 24, 26, 28, 30, 31, 29] # Celsius
ice_cream_sales = [120, 180, 250, 300, 400, 420, 380] # Dollars
```

```python
fig, ax1 = plt.subplots()

# First axis - Temperature:
ax1.plot(days, temperature, 'b-', marker = 'o')
ax1.set_xlabel("Day")
ax1.set_ylabel("Temperature (C)", color = 'b')

# Second axis - Sales
ax2 = ax1.twinx()
ax2.plot(days, ice_cream_sales, 'r--', marker = 's')
ax2.set_ylabel("Ice Cream Sales ($)", color = 'r')

plt.title("Temperature vs Ice Cream Sales")
plt.show()
```

### Demo 2: Line + Bar Chart with Twin Axes

```python
import numpy as np

months = ["Jan", "Feb", "Mar", "Apr", 'May', "Jun"]
revenue = [200, 250, 300, 280, 320, 350]
marketing_spend = [50, 80, 100, 90, 110, 120]

fig, ax1 = plt.subplots()

# Barchart for marketing spend:
ax1.bar(months, marketing_spend, color = "lightblue")
ax1.set_ylabel("Marketing Spend ($)", color = "blue")

# Twin axis for revenue:
ax2 = ax1.twinx()
ax2.plot(months, revenue, color = 'green', marker = 'o')
ax2.set_ylabel("Revenue ($)", color = 'green')


plt.title("Revenue vs Marketing Spend")
plt.show()
```

### Demo 3: line + Scatter with Twin Axes

```python
time = [1, 2, 3, 4, 5, 6]
website_visitors = [500, 600, 700, 800, 850, 900]
bounce_rate = [60, 55, 50, 52, 48, 45]
```

```python
fig, ax1 = plt.subplots()

# line chart for visitors:
ax1.plot(time, website_visitors, color = 'purple', marker = 'o')
ax1.set_ylabel("Website Visitors", color = 'purple')

# Scatter for bounce rate:
ax2 = ax1.twinx()
ax2.scatter(time, bounce_rate, color = "orange", s = 80)
ax2.set_ylabel("Bounce Rate (%)", color = 'orange')

plt.title("Website Visitors vs Bounce Rate")
plt.show()
```
