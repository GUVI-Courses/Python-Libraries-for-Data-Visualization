## Step 1: import matplotlib

```python
import matplotlib.pyplot as plt

# x-axis values (could be time, index, categories)
x = [1, 2, 3, 4, 5]

# y-axis values (corresponding measurements)
y = [10, 15, 12, 18, 20]

y2 = [8, 12, 15, 14, 22]


plt.plot(x, y, label = "Line 1", color = 'green', linestyle = '--',
        marker = 'o', markerfacecolor = 'blue', markersize = 8)

plt.plot(x, y2, label = "Line 2", color = 'red', marker = 's')
plt.title("Multiple Line Plot")

plt.xlabel("X-axis Label")

plt.ylabel("Y-axis Label")
plt.legend()
plt.grid(True)
plt.show()
```

## Barchart

```python
import matplotlib.pyplot as plt

products = ["Product A", "Product B", "Product C", "Product D"]

#Values in Y - axis:
sales = [25, 40, 30, 35]

plt.bar(products, sales, color = "skyblue", edgecolor = 'black')

plt.title("Sales of Products")
plt.xlabel("Products")
plt.ylabel("Sales")

# Add value labels on top of each bar:
for i, value in enumerate(sales):
    plt.text(i, value + 0.5, str(value), ha = 'center', va = 'bottom')


plt.show()
```

```python
plt.barh(products, sales, color = 'lightgreen', edgecolor = 'black')
plt.title("Sales of Products (Horizontal)")
plt.xlabel("Sales")
plt.ylabel("Products")
plt.show()
```

## Scatter plot

```python
import matplotlib.pyplot as plt

# Data:
hours_studied = [1, 2, 3, 4, 5, 6, 7, 8]
test_scores = [50, 55, 60, 65, 70, 72, 75, 80]

plt.scatter(hours_studied, test_scores, color = 'red', marker = 's', s = 100)
plt.title("Scatter Plot: Hours studied vs Test scores")
plt.xlabel("Hours Studied")
plt.ylabel("Test Scores")

plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data:
hours_studied_group1 = [1, 2, 3, 4]
scores_group1 = [50, 55, 60, 65]

hours_studied_group2 = [5, 6, 7, 8]
scores_group2 = [70, 72, 75, 80]

plt.scatter(hours_studied_group1, scores_group1, color = 'blue', marker = 'o', s = 100, label = "Group 1")
plt.scatter(hours_studied_group2, scores_group2, color = 'green', marker = 's', s = 100, label = "Group 2")

plt.title("Scatter Plot with Multiple Groups")
plt.xlabel("Hours Studied")
plt.ylabel("Test Scores")

plt.legend()
plt.grid(True)
plt.show()
```

## Histogram

```python
import matplotlib.pyplot as plt

scores = [55, 60, 65, 70, 75, 80, 85, 90, 95, 100,
         60, 65, 70, 75, 80, 85, 90, 95, 100, 100]


plt.hist(scores, bins = 10, color = 'orange', edgecolor = 'black')
plt.title("Distribution of Exam Scores")
plt.xlabel("Scores")
plt.ylabel("Number of Students")

plt.show()
```

```python
import matplotlib.pyplot as plt

scores_class1 = [55, 60, 65, 70, 75, 80, 85, 90, 95, 100]
scores_class2 = [60, 65, 70, 75, 80, 85, 90, 95, 100, 100]


plt.hist(scores_class1, bins = 5, color = 'blue', alpha = 0.5, label = 'Class 1', edgecolor = 'black')
plt.hist(scores_class2, bins = 5, color = 'green', alpha = 0.5, label = 'Class 2', edgecolor = 'black')

plt.title("Comparision of Exam Scores Between Two Classes")
plt.xlabel("Scores")
plt.ylabel("Number of Students")
plt.legend()

plt.show()
```

## Code Example - multiple Plots & Subplots

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample Data:
x = np.linspace(0, 10, 100) #100 points between 0 and 10
y1 = np.sin(x)  #Sine Curve
y2 = np.cos(x)  #Cosine Curve

categories = ["Apples", "Bananas", "Cherries", "Dates"]
values = [10, 15, 7, 12]

x_scatter = np.random.rand(50) * 10  # Random X values (0-10)
y_scatter = np.random.rand(50) * 100 # Random Y values (0-100)
scores = np.random.randint(50, 100, 30) # Random exam scores (50-100)


# --------------------------
# 1. Side-by-side plots (subplot with 1 row, 2 columns)
# --------------------------

fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (10, 4))

# First subplot: Line Plot:
ax1.plot(x, y1, label = "Sine", color = "blue")
ax1.plot(x, y2, label = "Cosine", color = "red", linestyle = '--')
ax1.set_title("Line Plot: Sine vs Cosine")
ax1.set_xlabel("X-axis")
ax1.set_ylabel("Y-axis")
ax1.legend()

# Second subplot: Barchart
ax2.bar(categories, values, color = "green")
ax2.set_title("Bar chart: Fruit Sales")
ax2.set_xlabel("Fruit")
ax2.set_ylabel("Quantity")

plt.tight_layout()
plt.show()

# -----------------------------------------
# 2. 2*2 Grid of Subplots
# -----------------------------------------

fig, axs = plt.subplots(2, 2, figsize = (12, 8))

# (1,1) Line Plot
axs[0, 0].plot(x, y1, color = "purple")
axs[0, 0].set_title("Sine Curve")

# (1, 2) BarChart
axs[0, 1].bar(categories, values, color = "orange")
axs[0, 1].set_title("Fruit Sales")

# (2,1) Scatter Plot
axs[1, 0].scatter(x_scatter, y_scatter, color = "teal", marker = 'o')
axs[1,0].set_title("Scatter: Random Data")
axs[1,0].set_xlabel("X Values")
axs[1, 0].set_ylabel("Y Values")

# (2, 2) Histogram
axs[1,1].hist(scores, bins = 5, color = "pink", edgecolor = "black")
axs[1,1].set_title("Histogram: Exam Scores")
axs[1,1].set_xlabel("Scores")
axs[1,1].set_ylabel("Frequency")

# Adjust Layout:
plt.suptitle("Multiple Plots & Subplots Example", fontsize = 16, y = 1.02)
plt.tight_layout()
plt.show()
```

## Annotation

```python
import matplotlib.pyplot as plt
```

```python
# Example Data:

plt.figure(figsize = (10,6), dpi = 120)

x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 25, 30]

plt.plot(x, y, marker = 'o')

# Simple text:
plt.text(3, 15, "Mid-Point", fontsize = 10, color = "blue")

# Annotation with Arrow:
plt.annotate("Peak Value",
            xy = (5, 30), xytext = (3.5, 28),
            arrowprops = dict(facecolor = 'red', shrink = 0.05))


plt.title("Annotation Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.style.use("ggplot")
plt.xticks(rotation = 45)
plt.grid(True, linestyle = '--', alpha = 0.7)

plt.show()
```

```python
## Advanced Customization

for style in ["default", "seaborn", "ggplot", "dark_background"]:
    plt.style.use(style)
    plt.plot(x, y, marker = 'o')
    plt.title(f"Style: {style}")
    plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate sample sales data
np.random.seed(42)
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", 
          "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
sales = np.random.randint(200, 600, size=12)

# Set figure size & style
plt.style.use("seaborn-v0_8")  # modern look
plt.figure(figsize=(12, 6), dpi=120)

# Plot line chart
plt.plot(months, sales, marker="o", color="blue", linewidth=2, label="Monthly Sales")

# Highlight max and min sales with annotations
max_month = months[np.argmax(sales)]
max_value = sales.max()
min_month = months[np.argmin(sales)]
min_value = sales.min()

plt.annotate(f"Peak: {max_value}", 
             xy=(max_month, max_value), 
             xytext=(8, max_value+5),
             arrowprops=dict(facecolor="red", arrowstyle="->"),
             fontsize=10, color="red")

plt.annotate(f"Low: {min_value}", 
             xy=(min_month, min_value), 
             xytext=(2, min_value-10),
             arrowprops=dict(facecolor="green", arrowstyle="->"),
             fontsize=10, color="green")

# Customize plot
plt.title(" Monthly Sales Trends (2025)", fontsize=16, fontweight="bold")
plt.xlabel("Month", fontsize=12)
plt.ylabel("Sales (in $)", fontsize=12)
plt.xticks(rotation=45)
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend()

# Add text directly on data points
for i, value in enumerate(sales):
    plt.text(i, value+10, str(value), ha="center", fontsize=9, color="black")

# Save figure
plt.tight_layout()
plt.savefig("sales_trends.png", dpi=150)
plt.show()
```
