## Boxplot

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(42)
data = np.random.normal(50, 10, 200) # avergae = 50, spread = 10

plt.boxplot(data)
plt.title("Boxplot Example: Exam Scores")
plt.ylabel("Scores")
plt.show()
```

```python
plt.figure(figsize = (8, 6)) #bigger figure
plt.boxplot(
data,
patch_artist = True,
    notch = True,
    vert = True,
    widths = 0.4,
    boxprops = dict(facecolor = "lightblue", color = 'blue', linewidth = 1.5),
    whiskerprops = dict(color = "blue", linewidth = 1.5),
    capprops = dict(color = "blue", linewidth = 1.5),
    medianprops = dict(color = "red", linewidth = 2),
    flierprops = dict(marker = 'o', markerfacecolor = "orange", markersize = 8, linestyle = 'none')
)

# Titles and labels:
plt.title("Customized Box Plot: Exam Scores", fontsize = 14, fontweight = "bold")
plt.ylabel("Scores", fontsize = 12)
plt.xlabel("Class A", fontsize = 12)

# Grid for clarity:
plt.grid(axis = "y", linestyle = '--', alpha = 0.7)

plt.show()
```

### Violin plot

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Sample Data:
np.random.seed(42)

data1 = np.random.normal(50, 10, 200) #Group1
data2 = np.random.normal(60, 8, 200)  #Group2
data3 = np.random.normal(55, 12, 200) #Group3


data = [data1, data2, data3]

# Create Violin Plot:
plt.violinplot(data)

plt.title("Basic Violin Plot (Matplotlib)")
plt.ylabel("Values")
plt.xticks([1,2,3],["Group A", "Group B", "Group C"]) # rename groups
plt.show()
```

### Customized Violin Plot

```python
plt.figure(figsize = (8, 6))

parts = plt.violinplot(
data,
    showmeans = True,   # Show mean line
    showmedians = True, # Show median marker
    showextrema = True  # Show min/ max
)

# Customize colors:
for pc in parts['bodies']:
    pc.set_facecolor("#1f77b4") # blue fill
    pc.set_edgecolor("black") # black border
    pc.set_alpha(0.6)   # transparency
    
# Customize other parts:
parts['cmedians'].set_color('red')
parts['cmeans'].set_color('green')
parts['cmaxes'].set_color('black')
parts['cmins'].set_color('black')

plt.xticks([1, 2, 3], ["Group A", "Group B", "Group C"])
plt.title("Customized Violin Plot (Matplotlib)", fontsize = 14, fontweight = "bold")
plt.ylabel("Values")
plt.show()
```

```python
### Side-by-side Comparisionn (Boxplot vs Violin Plot):

plt.figure(figsize = (12, 6))

# Left: Boxplot
plt.subplot(1,2,1)
plt.boxplot(data)
plt.title("Boxplot")
plt.xticks([1,2,3], ["Group A", "Group B", "Group C"])

# Right : Violin Plot
plt.subplot(1,2,2)
plt.violinplot(data, showmeans = True, showmedians = True)
plt.title("Violin Plot")
plt.xticks([1,2,3], ["Group A", "Group B", "Group C"])

plt.suptitle("Boxplot vs Violin PLot", fontsize = 16, fontweight = 'bold')
plt.show()
```

### Density Plot

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde
```

```python
data = np.random.normal(30, 10, 200) # mean = 30, std = 10, n = 200

# Perform Kernal Density Estimation
kde = gaussian_kde(data)
x = np.linspace(min(data), max(data), 100)
y = kde(x)

# plot:
plt.hist(data, bins = 15, density = True, alpha = 0.3, label = "Histogram")
plt.plot(x, y, color = 'red', label = "Density Plot (KDE)")
plt.title("Customer Age Distribution")
plt.legend()
plt.show()
```

### example 2- KDE

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import gaussian_kde
import seaborn as sns


# Load example dataset:
tips = sns.load_dataset("tips")
data = tips['total_bill']

# Compute KDE using scipy
kde = gaussian_kde(data)

# Create a range of values for smooth density curve:
x = np.linspace(min(data), max(data), 200)
y = kde(x)

# plot the density:
plt.fill_between(x, y, color = "skyblue", alpha = 0.5) # Shaded area
plt.plot(x, y, color = "blue", linewidth = 2) # KDE line

plt.title("Density Plot Example: Bill Amounts (Matplotlib)")
plt.xlabel("Total Bill")
plt.ylabel("Density")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import gaussian_kde
import seaborn as sns


# Load example dataset:
tips = sns.load_dataset("tips")
data = tips['total_bill']

# Compute KDE using scipy
kde = gaussian_kde(data, bw_method = 0.8)

# Create a range of values for smooth density curve:
x = np.linspace(min(data), max(data), 200)
y = kde(x)

# plot the density:
plt.fill_between(x, y, color = "skyblue", alpha = 0.5) # Shaded area
plt.plot(x, y, color = "blue", linewidth = 2) # KDE line
plt.hist(data, bins = 30, density = True, alpha = 0.25)

plt.title("Density Plot Example: Bill Amounts (Matplotlib)")
plt.xlabel("Total Bill")
plt.ylabel("Density")
plt.show()
```

### Contour Plot in matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Create grid of x and y values:
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
```

```python
# Function to visualize
Z = np.sin(np.sqrt(X ** 2 + Y ** 2))
```

```python
# Basic Contour plot:

plt.contour(X, Y, Z, levels = 10, cmap = 'viridis')
plt.colorbar(label = "Z Value")
plt.title("Basic Contour Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
plt.contourf(X, Y, Z, levels = 20, cmap = "plasma")
plt.colorbar(label = "Z Value")
plt.title("Filled Contour Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

### Customize Contour Plots

```python
contours = plt.contour(X, Y, Z, levels = 15, colors = "black", linewidth = 1)
plt.clabel(contours, inline = True, fontsize = 8) # add labels to lines
plt.contourf(X, Y, Z, levels = 15, cmap = "coolwarm", alpha = 0.7)

plt.colorbar(label = "Z Value")
plt.title("Customized Contour Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

### ML Example Code

```python
from sklearn.datasets import make_classification
from sklearn.linear_model import LogisticRegression
```

```python
# Generate sample data:
X, y = make_classification(n_features = 2, n_redundant = 0, n_informative = 2,
                          n_clusters_per_class = 1, random_state = 42)
model = LogisticRegression().fit(X, y)
```

```python
# Create grid for contour:

xx, yy = np.meshgrid(np.linspace(X[:,0].min()-1, X[:,0].max()+1, 100),
                     np.linspace(X[:,1].min()-1, X[:,1].max()+1, 100))

Z = model.predict_proba(np.c_[xx.ravel(), yy.ravel()])[:,1]
Z = Z.reshape(xx.shape)

plt.contour(xx, yy, Z, levels = 20, cmap = "RdBu", alpha = 0.7)
plt.scatter(X[:,0], X[:,1], c = y, edgecolor = "k", cmap = "RdBu", marker = 'o')
plt.title("Contour Plot: Logistic Regression Decision Boundary")
plt.show()
```
